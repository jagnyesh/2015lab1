```
import random

# Define the answer choices
ANSWER_CHOICES = {
    1: "Strongly Disagree",
    2: "Disagree",
    3: "Neutral",
    4: "Agree",
    5: "Strongly Agree"
}

# --- Question Bank ---
# You can expand this question bank with up to 100 questions or more.
# Ensure each question has an "id", "category", and "text".
# Try to have a mix of positively and negatively framed questions within each category
# to better test consistency.

QUESTIONS_DATA = [
    # Category: Ethics (General & Honesty)
    {"id": 1, "category": "Ethics", "text": "It is always necessary to adhere strictly to all company policies, without exception."},
    {"id": 2, "category": "Ethics", "text": "If I saw a coworker taking home office supplies for personal use, I would report it."},
    {"id": 3, "category": "Ethics", "text": "Sometimes, a small white lie is acceptable if it prevents hurting someone's feelings in a professional setting."},
    {"id": 4, "category": "Ethics", "text": "I believe that the end result can sometimes justify the means, even if it involves cutting minor corners."},
    {"id": 5, "category": "Ethics", "text": "Honesty is the most important principle in the workplace, above all else."},
    {"id": 6, "category": "Ethics", "text": "I would feel uncomfortable if a project's success depended on overlooking a minor ethical guideline."},
    {"id": 7, "category": "Ethics", "text": "It's okay to share confidential company information with close friends or family if they promise not to tell anyone."},
    {"id": 8, "category": "Ethics", "text": "I would take responsibility for a mistake even if it meant facing negative consequences."},
    {"id": 9, "category": "Ethics", "text": "If a client misunderstands something in a way that benefits the company, I would not rush to correct them."},
    {"id": 10, "category": "Ethics", "text": "I have never exaggerated my skills or experience on a resume or in an interview."},
    {"id": 11, "category": "Ethics", "text": "Maintaining fairness in all professional dealings is a top priority for me."},


    # Category: Ethics & User Safety (GHA focus: Child Safety, Vulnerable Populations)
    {"id": 12, "category": "Ethics & User Safety", "text": "Protecting vulnerable users should be the absolute highest priority, even if it impacts product features or timelines."},
    {"id": 13, "category": "Ethics & User Safety", "text": "If a feature has potential for misuse that could harm vulnerable populations, it should not be launched until all risks are fully mitigated."},
    {"id": 14, "category": "Ethics & User Safety", "text": "It is acceptable to collect more user data than strictly necessary if it might lead to future safety improvements."},
    {"id": 15, "category": "Ethics & User Safety", "text": "When faced with ambiguous content that might be harmful to children, I would err on the side of caution and restrict it."},
    {"id": 16, "category": "Ethics & User Safety", "text": "Company revenue targets should take precedence over minor user safety concerns that are unlikely to materialize."},
    {"id": 17, "category": "Ethics & User Safety", "text": "I am comfortable making decisions that restrict user freedom if it significantly enhances child safety online."},
    {"id": 18, "category": "Ethics & User Safety", "text": "A user-first philosophy means always prioritizing the well-being and safety of users, especially those most vulnerable."},
    {"id": 19, "category": "Ethics & User Safety", "text": "Transparency with users about how their data is used for safety purposes is important, but not if it compromises the effectiveness of safety measures."},
    {"id": 20, "category": "Ethics & User Safety", "text": "I believe technology companies have a profound ethical responsibility to actively protect their users from harm."},
    {"id": 21, "category": "Ethics & User Safety", "text": "If a safety protocol seems overly bureaucratic and slows down processes, it should be re-evaluated for efficiency, even if it's effective."},
    {"id": 22, "category": "Ethics & User Safety", "text": "Decisions about user safety should primarily be driven by data and evidence, rather than purely ethical considerations."},


    # Category: Communication
    {"id": 23, "category": "Communication", "text": "I am comfortable giving direct and constructive feedback to colleagues, even if it might be poorly received."},
    {"id": 24, "category": "Communication", "text": "I prefer to communicate important information via email rather than face-to-face or video call."},
    {"id": 25, "category": "Communication", "text": "When I disagree with someone, I tend to state my opinion clearly and assertively."},
    {"id": 26, "category": "Communication", "text": "I find it easy to adapt my communication style to different audiences."},
    {"id": 27, "category": "Communication", "text": "Active listening is as important as speaking clearly in effective communication."},
    {"id": 28, "category": "Communication", "text": "I sometimes avoid difficult conversations to maintain harmony within the team."},
    {"id": 29, "category": "Communication", "text": "I make an effort to ensure everyone in a meeting has a chance to speak."},
    {"id": 30, "category": "Communication", "text": "Written communication should always be formal and meticulously proofread."},
    {"id": 31, "category": "Communication", "text": "If I don't understand something, I will ask for clarification immediately, even if it interrupts the speaker."},
    {"id": 32, "category": "Communication", "text": "I believe it's important to share updates frequently, even if there isn't major progress to report."},
    {"id": 33, "category": "Communication", "text": "Non-verbal cues are a significant part of how I interpret communication."},

    # Category: Collaboration & Teamwork
    {"id": 34, "category": "Collaboration & Teamwork", "text": "I achieve my best work when I am part of a team."},
    {"id": 35, "category": "Collaboration & Teamwork", "text": "I am willing to compromise my own ideas for the sake of team consensus."},
    {"id": 36, "category": "Collaboration & Teamwork", "text": "When a team project is successful, I focus on the team's achievement rather than my individual contribution."},
    {"id": 37, "category": "Collaboration & Teamwork", "text": "I find it challenging to work with team members who have very different working styles from my own."},
    {"id": 38, "category": "Collaboration & Teamwork", "text": "I readily offer help to teammates, even if it's outside my direct responsibilities."},
    {"id": 39, "category": "Collaboration & Teamwork", "text": "In a team setting, it's better to avoid conflict to keep morale high."},
    {"id": 40, "category": "Collaboration & Teamwork", "text": "I believe diverse perspectives within a team lead to better outcomes."},
    {"id": 41, "category": "Collaboration & Teamwork", "text": "I am comfortable relying on others to complete their part of a team project."},
    {"id": 42, "category": "Collaboration & Teamwork", "text": "When working with cross-functional teams, bridging communication gaps is a key priority for me."},
    {"id": 43, "category": "Collaboration & Teamwork", "text": "I prefer clearly defined roles within a team project."},
    {"id": 44, "category": "Collaboration & Teamwork", "text": "Sharing credit for successes is important for team cohesion."},


    # Category: Work Structure & Organization
    {"id": 45, "category": "Work Structure & Organization", "text": "I thrive in highly structured work environments with clear rules and procedures."},
    {"id": 46, "category": "Work Structure & Organization", "text": "I prefer to plan my workday in detail rather than adapting as I go."},
    {"id": 47, "category": "Work Structure & Organization", "text": "Having a neat and organized workspace is very important for my productivity."},
    {"id": 48, "category": "Work Structure & Organization", "text": "I am good at prioritizing tasks and managing my time effectively."},
    {"id": 49, "category": "Work Structure & Organization", "text": "Too much process and bureaucracy can stifle creativity and efficiency."},
    {"id": 50, "category": "Work Structure & Organization", "text": "I prefer projects with clear, pre-defined goals and deliverables."},
    {"id": 51, "category": "Work Structure & Organization", "text": "I often find myself multitasking to get more done."},
    {"id": 52, "category": "Work Structure & Organization", "text": "I believe in following established processes even if I think there might be a faster way."},
    {"id": 53, "category": "Work Structure & Organization", "text": "Clear hierarchies and reporting lines make an organization more efficient."},
    {"id": 54, "category": "Work Structure & Organization", "text": "I am comfortable with a degree of ambiguity in my tasks and projects."},
    {"id": 55, "category": "Work Structure & Organization", "text": "Regular status meetings are essential for keeping projects on track."},

    # Category: Work Flexibility & Adaptability
    {"id": 56, "category": "Work Flexibility & Adaptability", "text": "I easily adapt to unexpected changes in project scope or deadlines."},
    {"id": 57, "category": "Work Flexibility & Adaptability", "text": "I prefer a consistent routine over frequent changes in my work tasks."},
    {"id": 58, "category": "Work Flexibility & Adaptability", "text": "I am open to learning new skills or technologies, even if they are outside my comfort zone."},
    {"id": 59, "category": "Work Flexibility & Adaptability", "text": "When priorities shift suddenly, I find it stressful and difficult to adjust."},
    {"id": 60, "category": "Work Flexibility & Adaptability", "text": "I am comfortable working with evolving safety challenges that require new approaches."},
    {"id": 61, "category": "Work Flexibility & Adaptability", "text": "I enjoy environments where I am constantly learning and adapting to new information."},
    {"id": 62, "category": "Work Flexibility & Adaptability", "text": "A flexible work schedule (e.g., choosing my own hours) is very important to me."},
    {"id": 63, "category": "Work Flexibility & Adaptability", "text": "I can quickly pivot to a new task or project when required."},
    {"id": 64, "category": "Work Flexibility & Adaptability", "text": "I find it easy to adopt new tools and processes in the workplace."},
    {"id": 65, "category": "Work Flexibility & Adaptability", "text": "Handling ambiguity well is one of my strengths."},
    {"id": 66, "category": "Work Flexibility & Adaptability", "text": "I prefer stability and predictability in my work environment over constant change."},

    # Category: Problem Solving & Analytical Thinking
    {"id": 67, "category": "Problem Solving & Analytical Thinking", "text": "I enjoy tackling complex problems that require deep analytical thought."},
    {"id": 68, "category": "Problem Solving & Analytical Thinking", "text": "When faced with a problem, I first try to gather all relevant data before considering solutions."},
    {"id": 69, "category": "Problem Solving & Analytical Thinking", "text": "I am skilled at breaking down complex issues into smaller, manageable parts."},
    {"id": 70, "category": "Problem Solving & Analytical Thinking", "text": "I often rely on my intuition when making quick decisions to solve a problem."},
    {"id": 71, "category": "Problem Solving & Analytical Thinking", "text": "I am comfortable making decisions even when there is significant ambiguity or incomplete information."},
    {"id": 72, "category": "Problem Solving & Analytical Thinking", "text": "I tend to think outside the box to find innovative solutions."},
    {"id": 73, "category": "Problem Solving & Analytical Thinking", "text": "When problem-solving, I consider multiple potential solutions and their consequences."},
    {"id": 74, "category": "Problem Solving & Analytical Thinking", "text": "I find it more effective to use proven methods for problem-solving rather than experimenting with new ones."},
    {"id": 75, "category": "Problem Solving & Analytical Thinking", "text": "Identifying the root cause of a problem is more important than finding a quick fix."},
    {"id": 76, "category": "Problem Solving & Analytical Thinking", "text": "I am persistent in trying to solve a problem, even when it's very challenging."},
    {"id": 77, "category": "Problem Solving & Analytical Thinking", "text": "I am good at anticipating potential problems before they arise."},


    # Category: Responsibility & Decision Making
    {"id": 78, "category": "Responsibility & Decision Making", "text": "I am comfortable taking full responsibility for my decisions, even if they turn out to be wrong."},
    {"id": 79, "category": "Responsibility & Decision Making", "text": "I prefer to have clear guidance from a superior before making important decisions."},
    {"id": 80, "category": "Responsibility & Decision Making", "text": "When making decisions, I primarily consider the potential impact on the company's bottom line."},
    {"id": 81, "category": "Responsibility & Decision Making", "text": "I am quick to make decisions, even under pressure."},
    {"id": 82, "category": "Responsibility & Decision Making", "text": "I believe in consulting with team members before making a decision that affects them."},
    {"id": 83, "category": "Responsibility & Decision Making", "text": "Once I make a decision, I rarely second-guess it."},
    {"id": 84, "category": "Responsibility & Decision Making", "text": "I take initiative and often make decisions without being asked."},
    {"id": 85, "category": "Responsibility & Decision Making", "text": "It's important to consider all possible risks before making a significant decision."},
    {"id": 86, "category": "Responsibility & Decision Making", "text": "I am willing to make unpopular decisions if I believe they are right for the organization."},
    {"id": 87, "category": "Responsibility & Decision Making", "text": "I prefer when decisions are made by consensus rather than by a single individual."},
    {"id": 88, "category": "Responsibility & Decision Making", "text": "I hold myself accountable for meeting my commitments and deadlines."},

    # Category: Leadership
    {"id": 89, "category": "Leadership", "text": "I naturally take on a leadership role in group settings."},
    {"id": 90, "category": "Leadership", "text": "Motivating others comes easily to me."},
    {"id": 91, "category": "Leadership", "text": "I believe a good leader should primarily focus on achieving results."},
    {"id": 92, "category": "Leadership", "text": "I am comfortable delegating tasks to others."},
    {"id": 93, "category": "Leadership", "text": "A leader's main role is to provide clear direction and vision."},
    {"id": 94, "category": "Leadership", "text": "I prefer to lead by example rather than by giving orders."},
    {"id": 95, "category": "Leadership", "text": "Empowering team members to make their own decisions is a key leadership trait."},
    {"id": 96, "category": "Leadership", "text": "I find it challenging to manage conflicts within a team I am leading."},
    {"id": 97, "category": "Leadership", "text": "Good leaders are born, not made."},
    {"id": 98, "category": "Leadership", "text": "Providing regular feedback and coaching is crucial for effective leadership."},
    {"id": 99, "category": "Leadership", "text": "A leader should be able to make tough decisions that may not be popular with the team."},
    {"id": 100, "category": "Leadership", "text": "I am more of a follower than a leader in most situations."}
]


def display_welcome_message():
    """Displays a welcome message to the user."""
    print("-------------------------------------------------")
    print("Welcome to the Consistency Practice Test!")
    print("-------------------------------------------------")
    print("You will be presented with a series of statements.")
    print("For each statement, please choose the option that best reflects your view.")
    print("The goal is to practice answering consistently based on your personal values and approach.")
    print(f"\nThere are {len(QUESTIONS_DATA)} questions in this test.")
    input("\nPress Enter to begin...")
    print("-" * 50)


def get_user_answer():
    """Gets and validates the user's answer for a question."""
    print("\nPlease choose an answer (1-5):")
    for key, value in ANSWER_CHOICES.items():
        print(f"{key}. {value}")

    while True:
        try:
            choice = int(input("Your choice: "))
            if choice in ANSWER_CHOICES:
                return ANSWER_CHOICES[choice]
            else:
                print("Invalid choice. Please enter a number between 1 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number.")


def run_test(questions):
    """Runs the test, presenting questions and collecting answers."""
    user_answers = []
    shuffled_questions = random.sample(questions, len(questions)) # Shuffle questions

    for i, question_data in enumerate(shuffled_questions):
        print(f"\nQuestion {i + 1}/{len(shuffled_questions)}")
        print(f"Category: {question_data['category']}")
        print(f"Statement: {question_data['text']}")

        answer = get_user_answer()
        user_answers.append({
            "id": question_data['id'], # Store original ID for stable reference
            "category": question_data['category'],
            "text": question_data['text'],
            "answer": answer
        })
        print("-" * 50)
    return user_answers


def display_review(user_answers):
    """Displays the user's answers, grouped by category for review."""
    print("\n-------------------------------------------------")
    print("Test Completed! Review Your Answers:")
    print("-------------------------------------------------")

    answers_by_category = {}
    for ans in user_answers:
        category = ans['category']
        if category not in answers_by_category:
            answers_by_category[category] = []
        answers_by_category[category].append(ans)

    # Sort answers within each category by original question ID for consistent review order
    for category in answers_by_category:
        answers_by_category[category].sort(key=lambda x: x['id'])


    for category, answers_in_cat in answers_by_category.items():
        print(f"\n--- Category: {category} ---")
        for i, ans_data in enumerate(answers_in_cat):
            # Find original question number if needed, though ID is more stable
            print(f"  Q (ID {ans_data['id']}): {ans_data['text']}")
            print(f"  Your Answer: {ans_data['answer']}\n")
        print("-" * 30)

    print("\nReview complete. This helps you check for consistency in your responses within each category.")


def main():
    """Main function to run the consistency practice test."""
    display_welcome_message()
    # If you want to take a subset of questions, you can slice QUESTIONS_DATA here
    # e.g., questions_to_ask = QUESTIONS_DATA[:50] for a 50-question test
    questions_to_ask = QUESTIONS_DATA

    if not questions_to_ask:
        print("No questions defined. Please add questions to QUESTIONS_DATA.")
        return

    user_responses = run_test(questions_to_ask)
    display_review(user_responses)


if __name__ == "__main__":
    main()
```
